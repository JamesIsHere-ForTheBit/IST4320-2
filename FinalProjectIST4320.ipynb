```python
#Imports
import tkinter as tk
from tkinter import ttk
import random
import sqlite3 #I know our readings stopped at Chapter 14 for Module 7, but I downloaded the PDF
#and realised in chapter 15 there is sqlite3 that allows saving to a local database file, so I am using that (Sorry for going
#ahead of what was assigned!) to fulfill the requirements of the final

#Also I have some SQL experience in another class, so I hope it is okay I brought some of that over to make this assignment work
#and I also used chapter 15 of the online book as a refresher


#Snack Choices
Yummy_Snacks = {
    'Savory': [ #I did savory first because those are my favorite haha
        'Chicken Nuggets',
        'Sushi',
        'Jerky',
        'Salami',
        'String Cheese'
    ],
    'Sweet': [
        'Applie Pie Slice',
        'Vanilla Icecream',
        'Oatmeal Raisin Cookie',
        'Costco Cold Brew', #It's like drining caffeinated icecream, can recommend
        'Chocolate Lava Cake'
    ],
    'Drinkies': [
        'Water',
        'Coffee',
        'Energy Drink',
        'Soda',
        '"Post-Final Exams Beverage"'
        
    ],
    'Healthier Options': [
        'Apple',
        'Salad',
        'Nuts',
        'Banana',
        'A singular glass of water'
    ]}

#Database
#Apologies if any of the database stuff is also a mess, and to be transparent I had to YouTube some videos on it
conn = sqlite3.connect('yummysnacks.db')
c = conn.cursor()

#Below table name was wrong, so changed to snacks
c.execute('''
    CREATE TABLE IF NOT EXISTS snacks ( 
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        category TEXT,
        snack TEXT
    )
''')
conn.commit()
current_snack = ''

#Custom Function
def pick_snack(): #changed name of this line to be easier to read from how it was before
    global current_snack
    category = category_var.get()
    snacks = Yummy_Snacks.get(category, [])
    
#Deleted below section to redo, could not make it work the way I had it...
#Also indentation of all of the below lines was wrong, fixed
    if snacks:
        current_snack = random.choice(snacks)
        result_label.config(text=current_snack)
    else:
        result_label.config(text='Please select a category!')

def save_snack():
    if current_snack == '':
        status_label.config(text='Please pick a snack!')
        return #Had return on the wrong indentation, fixed

    category = category_var.get()
    c.execute('INSERT INTO snacks (category, snack) VALUES (?, ?)',
              (category, current_snack))
    conn.commit()
    status_label.config(text='Snack is saved!') #Put texts instead of text, fixed

#Tkinter
window = tk.Tk()
window.title=('The Cool Snack Logger 9000')

title_label = tk.Label(window, text = 'Cool Snack Logger 9000 App', font = ('Arial', 20)) #Forgot '' around font, fixed
title_label.pack(pady = 10)

category_var = tk.StringVar()

category_label = tk.Label(window, text='Please Choose a category:')
category_label.pack()

category_box = ttk.Combobox(window, textvariable=category_var,
                            state = 'readonly', #Thought this was cool, didn't know how to implement this before this project!
                            values = list(Yummy_Snacks.keys()))
category_box.pack(pady = 5)

pick_button = tk.Button(window, text = 'Please pick a Random Snack!', command=pick_snack)
pick_button.pack(pady = 10)

result_label = tk.Label(window, text = 'Your snack will be displayed here!')
result_label.pack(pady = 10)

save_button = tk.Button(window, text = 'Save the Snack!', command = save_snack)
save_button.pack(pady = 5)

status_label = tk.Label(window, text = '', fg='red') #Red is my favorite color so I like to use that lol
status_label.pack(pady = 5)

window.mainloop()
conn.close()

#EUREKA it works!!!!
```
